```python
#!/usr/bin/env python
import pycurl, cStringIO
import string
import collections
import time
import csv
from itertools import groupby
import pandas
import scipy.io
import scipy.io
import numpy
```

```python
buf = cStringIO.StringIO()
data = {
    'token': 'E1F83E6A50C2F520800A49294B84C1B2',

    'content': 'report',
    'format': 'csv',
    'report_id': '22120',
    'rawOrLabel': 'raw',
    'rawOrLabelHeaders': 'raw',
    'exportCheckboxLabel': 'false',
    'returnFormat': 'csv'
}

ch = pycurl.Curl()
ch.setopt(ch.URL, 'https://redcap.stanford.edu/api/')
ch.setopt(ch.HTTPPOST, data.items())
ch.setopt(ch.WRITEFUNCTION, buf.write)
ch.perform()
ch.close()
patient_clinical_scores=buf.getvalue()
buf.close()
```
