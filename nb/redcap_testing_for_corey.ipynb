```python
#!/usr/bin/env python
import pycurl, cStringIO
import scipy.io
import json
import numpy
```

```python
buf = cStringIO.StringIO()
data = {
    'token': 'E1F83E6A50C2F520800A49294B84C1B2',
    'content': 'record',
    'format': 'json',
    'type': 'flat',
    'forms[0]': 'clinical_surveys',
    'rawOrLabel': 'raw',
    'rawOrLabelHeaders': 'raw',
    'exportCheckboxLabel': 'false',
    'exportSurveyFields': 'false',
    'exportDataAccessGroups': 'false',
    'returnFormat': 'json'
}
ch = pycurl.Curl()
ch.setopt(ch.URL, 'https://redcap.stanford.edu/api/')
ch.setopt(ch.HTTPPOST, data.items())
ch.setopt(ch.WRITEFUNCTION, buf.write)
ch.perform()
ch.close()
redcap_clinical_survey_output=buf.getvalue() #set buff.getvalue() to a variable
buf.close()
```
